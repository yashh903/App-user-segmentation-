```python
x=df.drop("Status",axis=1)
```

```python
x
```

```python
from sklearn.preprocessing import MinMaxScaler
for i in x.columns:
    MinMaxScaler(i)
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans=KMeans(n_clusters=3,init="k-means++")
y=kmeans.fit_predict(x)
x["Segments"] = y
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(x,y)
```

```python
x["Segments"].value_counts()
```

```python
x["Segments"] = x["Segments"].map({0: "Retained", 1: 
    "Churn", 2: "Needs Attention"})
```

```python
x
```

```python
for i in list(x["Segments"].unique()):
    sns.scatterplot(x=x[x["Segments"]==i]['Last Visited Minutes'],y=x[x["Segments"] == i]['Average Spent on App (INR)'],label= i)
```
